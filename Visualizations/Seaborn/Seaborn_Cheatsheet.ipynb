### Seaborn
Seaborn is a visualization libary that wraps matplotlib and does not ddo any actual plotting itself.
Seaborn makes Beautiful plots and has many types of plots that are not available from matplotlib or pandas

```python
#Installing the Seaborn Libary
#!pip install seaborn
```

```python
#Importing Seaborn --You need to import matplotlib first --as seaborn wraps around matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To get names of inbuilt dataset in seaborn python
sns.get_dataset_names()
```

```python
#Loading 'Tips' inbuilt dataset
tips = sns.load_dataset('tips')
tips
```

### Univariate → “one variable” data visualization
It means one variable analysis/visualization.When basically taking one variable under study is called univariate/one variable analysis

```python
#strip plot--1st-- When jitter=True

#When jitter=True,it plots for each observations a different data points

sns.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#strip plot--2nd-- When jitter=False

#When jitter=False,it plots for each observations a single data points

sns.stripplot(y= 'tip', data=tips,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#stripplot-- 3rd-- Grouping with stripplot()-- jitter=True

#having 2 variables, one on x-axis and one on y-axis

sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#stripplot-- 4th-- Grouping with stripplot()--jitter=false

#having 2 variables, one on x-axis and one on y-axis, 

sns.stripplot(x='day', y='tip', data=tips,size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot --1st--(Two parameter, day & tip) 

#it is used to get the represention of maximum data points of a particular observation

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarmplot --2nd -- (3 Parameter, days, tips, sex)
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarmplot --3rd-- changing orientation of 2nd
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

### Box and Violin plot


```python
#Box plot in Seaborn
#plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python
#Violin plot
#plt.subplot(1,2,2) 
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Combining plots --1st
#Plotting both Boxplot & Violinplot Simultaneously in different coln of subplots
plt.subplot(1,2,1)  #--(Row=1, coln=2, 1=boxplot will go in 1st coln of the subplot)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2) #--(Row=1, coln=2, 2=boxplot will go in 2nd coln of the subplot)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#Combining plots --2nd
#Plotting both Voilinplot & stripplot Simultaneously in same coln of subplots
sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgrey')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

### Bivariate → “two variables” data visualization
It means Two variable analysis/visualization.When basically taking Two variable under study is called univariate/one variable analysis

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot -- kind='kde'
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
#Pair plot --1st--to study relation between all variable of dataset 

sns.pairplot(tips)
plt.show()
```

```python
#Pairplot --2nd--to study relation between all variable of dataset with one variables as main
sns.pairplot(tips, hue='sex')
plt.show()
```
