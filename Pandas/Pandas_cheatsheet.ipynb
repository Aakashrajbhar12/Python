#### Pandas Data Structures --Series, Data Frames


```python
#Series -- 1D Structure
s=pd.Series([3,-5,7,4])  #Creating series having default index and dtype
print(s)
s1=pd.Series([3,-5,7,4], index=['a','b','c','d'], dtype=float) #Creating Series with Own Index and dtype
print(s1)
```

#### Data Frames -- 2D Structure, which Column & row like structure,having/storing Different data types in each

#### Data Frames data types --object(string),Int64(int),Float64(Float),Datatime64 etc.

```python
#Creating a Data Frame

data = {'Country': ['Belgium', 'India', 'Brazil'], 
        'Capital': ['Brussels', 'New Delhi', 'Brasília'],
        'Population': [11190846, 1303171035, 207847528]}   #First Creating Data 

df = pd.DataFrame(data, columns=['Country', 'Capital', 'Population']) #Creating Data frame
print(df)
```

```python
#Checking dtype of data frame  --Both of these can be use to find out the dtype
df.info()
df.dtypes   
```

#### Viewing Data and Data Frames --As per our requirement by using Data Frame Methods



```python
#See the top & bottom rows of the frame --head() & tail() methods

df1.head(30)   #Will give the number specified, gives the first 30 records
df1.head()     #by default gives 5 records/Data,gives the first 5 records

df1.tail(10)   #Will give the number specified, gives the last 10 records
df1.tail()     #by default gives 5 records/Data,gives the last 5 records
```

```python
#Viewing Only the index, columns, and the Values/data separately of the data frames,
#By using Data frames attributes such as: index,columns,values

#to view index 
df1.index    #Describes the index --starting and stop position & step count

#to view columns
df1.columns      #gives the name of the column
df1.columns[3]   #gives the column of the specified  column index

#to view Values
df1.values      #gives the entire values in an array
df1.values[73]  #gives the values of the specified row index
```

```python
#To Know the Structure of Data frame
#By Using Data frames Attributes such as: dtypes,axes,ndim,size,shape.

#to know dtypes of dataframe
df1.dtypes                 #gives dtypes for each columns
df1['salary'].dtypes       #gives dtypes of a particular colmn

#to know axes of dataframe --row & Col index
df1.axes                                   #gives rangeindex of row,index name of col

#to know the dimensions of data frame
df1.ndim                            #gives no. of dimension

#to know the size of data frame --row*Column
df1.size                                   #gives size by multiplying Row*Column

#to know the Shapes of data frame 
df1.shape                       #Gives no of row and column, (rows,column)
```

```python
#Stats / Summary Methods of Data frames

df1.max()         #to get max values
df1.min()         #to get min values
df1.mean()        #to get mean
df1.median()      #to get median
df1.std()         #to get standard deviation

#to get all stats value at once
df1.describe()                # gives all statistics values

#to have random sample from data frames
df1.sample()                     #gives one random samples
df1.sample(10)                   #gives the specified no of random samples i.e:10

#to Drop NA values from data frames --eliminate the entire rows
df1.dropna()
```

```python
#To get a particular column with values
df1.salary            #gives the column in a series structures
df1['salary']         #gives the column in a series structures
df1[['salary']]       #gives the coln in a Data frames structures

#Note: [] -- gives a series,[[]]--gives a data frame
```

```python
#Sorting of data in data frames

#Transposing the dataframe
df1.T                    #Gives a Transpose Data frame

#Sorting by Index 
df1.sort_index()        # sorts as per index

#Sorting by a column/values --sort_values()
df1.sort_values(by=['salary'], ascending=False)     # sorts by column i.e salary
```

```python
#Grouping of data in data frames --groupby()
#spliting of data into groups based on some criteria

grouped=df1.groupby(['rank']).mean()   # groupby ranl column and mean of all numberic data, this is all chaining two function
print(grouped)
type(grouped)
```

```python
#Filtering

df1.salary>12000                #gives a series 1D structures

#to Subset the data we can apply boolean indexing 
df1[df1['salary']>120000]       #gives a 2D data frame structure, this is a boolean indexing filter
```

```python
#Slicing --By position-- iloc[row,coln]

df1.iloc[0:10,0:3]    #will give row from 0 to 9, coln 0 to 3
df1.iloc[:,4]         #Will give all row and 4th coln
df1.iloc[0:2,]        #will give row 0,1 and all coln because not specified
```

```python
#Slicing --By label --loc[row,['coln label']]

df1.loc[0,['rank']]                 #give rank at index 0 of row
df1.loc[1,['rank','salary']]        #give rank & salary at index 1 of row
```
